```python
import pandas as pd
```

```python
df = pd.read_csv('RickAndMortyScripts.csv')
df.line=df.line.str.lower()
df.drop(['index','season no.','episode no.','episode name'], axis=1, inplace = True)
df
```

```python
for i in df:
    df['line'] = df['line'].str.replace(r'[^\w\s]+', '')
df
```

```python
with open('stopwords.txt') as file:
    stopWords = [word.rstrip() for word in file]

def getTopRMWordsMorty(df, stopWords, numWords):
    RMWords = {}
    for idx in range(len(df)):
        if df.iat[idx, 0] == 'Morty':
         line = df.iat[idx, 1]
         line = line.split()
         for word in line:
            if word not in stopWords:
                if word not in RMWords:
                    RMWords[word] = 1
                else:
                    RMWords[word] = RMWords[word] + 1

    RMWords = sorted(RMWords.items(), key=lambda x:x[1], reverse=True)
    return dict(RMWords[:numWords])

resultMort = getTopRMWordsMorty(df, stopWords, 5)
resultMort
```

```python
def getTopRMWordsRick(df, numWords):
    RMWords = {}
    for idx in range(len(df)):
        if df.iat[idx, 0] == 'Rick':
         line = df.iat[idx, 1]
         line = line.split()
         for word in line:
            if word not in stopWords:
                if word not in RMWords:
                    RMWords[word] = 1
                else:
                    RMWords[word] = RMWords[word] + 1

    RMWords = sorted(RMWords.items(), key=lambda x:x[1], reverse=True)
    return dict(RMWords[:numWords])

resultRick = getTopRMWordsRick(df, 5)
print(resultRick)
```

```python
def getTopRMWordsOther(df, numWords):
    RMWords = {}
    for idx in range(len(df)):
        if df.iat[idx, 0] != 'Rick''Morty':
         line = df.iat[idx, 1]
         line = line.split()
         for word in line:
            if word not in stopWords:
                if word not in RMWords:
                    RMWords[word] = 1
                else:
                    RMWords[word] = RMWords[word] + 1

    RMWords = sorted(RMWords.items(), key=lambda x:x[1], reverse=True)
    return dict(RMWords[:numWords])

resultOther = getTopRMWordsOther(df, 5)
print(resultOther)
```

```python
topListMorty=['rick','oh',"know","im","like"]
topListRick=['morty','gonna','youre','get','know']
topListOther=['morty','im','oh','rick','know']

def listing(topListMorty):
    ret = []
    for i in df.line:
        isp = i.split()
        retapp=[]
        for j in topListMorty:
            for x in isp:
                if j.lower() == x.lower():
                    retapp.append(1)
                    break
                if j.lower() != x.lower() and x == isp[-1]:
                    retapp.append(0)
                    break
        ret.append(retapp)
    return ret
vectorsMorty = listing(topListMorty)
print('vectors morty')
print(vectorsMorty)

def listing(topListRick):
    ret = []
    for i in df.line:
        isp = i.split()
        retapp=[]
        for j in topListRick:
            for x in isp:
                if j.lower() == x.lower():
                    retapp.append(1)
                    break
                if j.lower() != x.lower() and x == isp[-1]:
                    retapp.append(0)
                    break
        ret.append(retapp)
    return ret
vectorsRick = listing(topListRick)
print('vectors rick')
print(vectorsRick)

def listing(topListOther):
    ret = []
    for i in df.line:
        isp = i.split()
        retapp=[]
        for j in topListOther:
            for x in isp:
                if j.lower() == x.lower():
                    retapp.append(1)
                    break
                if j.lower() != x.lower() and x == isp[-1]:
                    retapp.append(0)
                    break
        ret.append(retapp)
    return ret
vectorsOther = listing(topListMorty)
print('vectors other')
print(vectorsOther)

```

```python
df['MortyVector'] = listing(topListMorty)
df['RickVector'] = listing(topListRick)
df['OtherVector'] = listing(topListOther)
df
```

```python
length = len(df.index)
splitPoint = int(length * 0.7)
train = df[0:splitPoint]
test = df[splitPoint:length]
```

```python
def makeProbTableMorty(train):
  feature_records = []
  for idx in range(len(train)):
    feature_vector = []
    ln = train.iat[idx, 1]
    nm = train.iat[idx, 0]
    for word in resultMort.keys():
      feature_vector.append(1 if word in ln else 0)
    feature_vector.append(nm)
    feature_records.append(feature_vector)
  feature_df = pd.DataFrame(feature_records)
  columns = list(resultMort.keys())
  columns.append('nm')
  feature_df.columns = columns
  return feature_df

probTableMorty = makeProbTableMorty(train)
probTableMorty
```

```python
def makeProbTableRick(train):
  feature_records = []
  for idx in range(len(train)):
    feature_vector = []
    ln = train.iat[idx, 1]
    nm = train.iat[idx, 0]
    for word in resultRick.keys():
      feature_vector.append(1 if word in ln else 0)
    feature_vector.append(nm)
    feature_records.append(feature_vector)
  feature_df = pd.DataFrame(feature_records)
  columns = list(resultRick.keys())
  columns.append('nm')
  feature_df.columns = columns
  return feature_df

probTableRick = makeProbTableRick(train)
probTableRick
```

```python
def makeProbTableOther(train):
  feature_records = []
  for idx in range(len(train)):
    feature_vector = []
    ln = train.iat[idx, 1]
    nm = train.iat[idx, 0]
    for word in resultOther.keys():
      feature_vector.append(1 if word in ln else 0)
    feature_vector.append(nm)
    feature_records.append(feature_vector)
  feature_df = pd.DataFrame(feature_records)
  columns = list(resultOther.keys())
  columns.append('nm')
  feature_df.columns = columns
  return feature_df

probTableOther = makeProbTableOther(train)
probTableOther
```

```python
p_rick = probTableRick[probTableRick.nm == 'Rick'].nm.count() / probTableRick.shape[0]
p_morty = probTableMorty[probTableMorty.nm == 'Morty'].nm.count() / probTableMorty.shape[0]
p_other = probTableOther[probTableOther.nm != 'Rick''Morty'].nm.count() / probTableOther.shape[0]

print(p_rick, p_morty, p_other)
```

```python
def getProbWordMorty(nm, probTableMorty):
  df = probTableMorty[probTableMorty.nm == nm]
  counts = df[resultMort.keys()].sum(axis=0) # compute # of 1s
  word_probs = counts / df.shape[0]
  word_probs = word_probs.to_dict()
  return word_probs

# each of these dictionaries has key (feature word): value (p(feature=1|label))
notMorty_probs = getProbWordMorty("Rick", probTableMorty)
Morty_probs = getProbWordMorty("Morty", probTableMorty)
```

```python
def getProbWordRick(nm, probTableRick):
  df = probTableRick[probTableRick.nm == nm]
  counts = df[resultRick.keys()].sum(axis=0) # compute # of 1s
  word_probs = counts / df.shape[0]
  word_probs = word_probs.to_dict()
  return word_probs

# each of these dictionaries has key (feature word): value (p(feature=1|label))
notRick_probs = getProbWordRick("Morty", probTableRick)
Rick_probs = getProbWordRick("Rick", probTableRick)
```

```python
def predict(feature_vector):
  '''
  Performs the Naive Bayes calculation P(y | x_i) = P(x_i) * P(y)
  Returns the label with the higher probability.
  '''
  Morty_prob = p_morty
  notMorty_prob = p_rick
  feature_words = list(resultMort.keys())
  for idx in range(len(feature_vector)):
    current_word = feature_words[idx]
    if feature_vector[idx] == 1:
      Morty_prob = Morty_prob * Morty_probs[current_word]
      notMorty_prob = notMorty_prob * notMorty_probs[current_word]
    else:
      # to compute p(x_i=0|label), we just subtract that
      # probably from 1: so p(x_i=0|label) = 1 - p(x_i=1|label).
      # Why do you think that is? Would that work with multiclass
      # naive bayes?
      Morty_prob = Morty_prob * (1-Morty_prob[current_word])
      notMorty_prob = notMorty_prob * (1-notMorty_prob[current_word])
  print(f"P(notMorty) = {notMorty_prob:.2f}")
  print(f"P(Morty) = {Morty_prob:.2f}")
  if notMorty_prob > Morty_prob:
    return 'notMorty'
  else:
    return 'Morty'
```

```python
# Calculates the percent of messages correctly labeled of the test dataset
correct = 0
for idx, (name, line, feature_vector) in train.iterrows():
  print(line)
  predicted_label = predict(feature_vector)
  print(name, predicted_label)
  if name == predicted_label:
    correct += 1
print(correct / test.shape[0])
```
