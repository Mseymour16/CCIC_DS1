```python
def printAllPage():
    pageNum = 1
    pages=[]
    while True:
        url = 'https://api.themoviedb.org/3/movie/top_rated?api_key=87983fbe976988a9e22330ebf030a1ae&language=en-US&page={0}'.format(pageNum)
        req = requests.get(url)
        if not req.ok:
            break
        pages = pages + req.json()['results']
        pageNum += 1
        
    return pages
```

```python
listOfMovies = printAllPage()
```

```python
df = pd.DataFrame(listOfMovies)
df
```

```python
df.drop(['backdrop_path'], axis=1, inplace=True)
df.drop(['adult','poster_path','overview','video','vote_count','release_date','original_language','original_title'], axis=1, inplace=True)
df
```

```python
def getRevenue():
    revenues = []
    for index, movie_entry_row in df.iterrows():
        
        url = 'https://api.themoviedb.org/3/movie/{0}?api_key=87983fbe976988a9e22330ebf030a1ae&language=en-US'.format(movie_entry_row["id"])
        req = requests.get(url)
        
        temp = req.json()['revenue']
        revenues.append(temp)
        
    return revenues

revenues = getRevenue()
revenues
```

```python
len(revenues)
```

```python
dfTemp = pd.DataFrame(revenues)
dfTemp.rename(columns={0 : 'revenue'}, inplace = True)
```

```python
data = pd.concat([df, dfTemp], axis=1)
data.head()
```

```python
data.sort_values("revenue",ascending = False, inplace = True)
data.drop(data.loc[data['revenue']==0].index, inplace = True)
data
```

```python
top10movies = data.head(10)

def getDirect():
    director = []
    for index, movie_entry_row in top10movies.iterrows():
        
        url = 'https://api.themoviedb.org/3/movie/{0}/credits?api_key=87983fbe976988a9e22330ebf030a1ae&language=en-US'.format(movie_entry_row["id"])
        req = requests.get(url)
        
        singleDirectors = []
                                                                                                         
        crewlist = req.json()['crew']
        for crewMember in crewlist:
            if crewMember['job']=='Director':
                singleDirectors.append(crewMember['name'])
        director.append(singleDirectors)
    return director
                                                                                                                              
top10directors = getDirect()
top10directors
```

```python
directorList = top10directors.copy()

for i in range(0, len(directorList)):
    if len(directorList[i]) > 1:
        directorList[i] = [directorList[i][0] + ", " + directorList[i][1]]

director_df = pd.DataFrame(directorList)
director_df
```
